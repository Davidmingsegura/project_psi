```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.cm as cm
```

```python
df0 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/raw_alumina.csv')
df1 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/Pt_al2o3_raw.csv')
df2 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/Pt_al2o3_200c.csv')
df3 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/Pt_al2o3_300C.csv')
df4 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/Pt_al2o3_400C.csv')
#df5 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/Pt_al2o3_500C.csv')
df6 = pd.read_csv('/Users/davidsegura/git/project_Ib_PSI/Carboncapture/XRD data/Pt_al2o3_600C.csv')
```

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(2,3, figsize=(18, 12))
axes = axes.flatten() 

titles = ['Raw Al2O3', '1% Pt/Al2O3 raw', '1% Pt/Al2O3 calcinated at 200C', '1% Pt/Al2O3 calcinated at 300C', '1% Pt/Al2O3 calcinated at 400C', '1% Pt/Al2O3 calcinated at 600C' ]
columns_x = ['2 theta', '2 theta', '2 theta', '2 theta', '2 theta', '2 theta']
columns_y = ['Intensity', 'Intensity', 'Intensity', 'Intensity', 'Intensity', 'Intensity']
dfs = [df0, df1, df2, df3, df4, df6]

plt.suptitle('XRD diffraction patterns', fontsize=20, weight='bold')

for i, ax in enumerate(axes):
    sns.lineplot(data=dfs[i], x=columns_x[i], y=columns_y[i], ax=ax)
    ax.set_title(titles[i], fontsize=16, weight='bold')
    ax.set_xlabel(f'{columns_x[i]} []', fontsize=14)
    ax.set_ylabel(f'{columns_y[i]} [a.u.]', fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=12)
    ax.grid(True)

plt.tight_layout()

plt.subplots_adjust(hspace=0.2, wspace=0.2)

#plt.savefig('six_plots.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
num_spectra = len(dfs)

colors = cm.get_cmap('coolwarm', num_spectra)

fig, ax = plt.subplots(figsize=(10, 6))

for i, df in enumerate(dfs):
    if i > 0:
        ax.plot(df['2 theta'], df['Intensity'], label=f'Spectrum {i+1}', color=colors(i))

ax.set_xlabel('2 theta []')
ax.set_ylabel('Intensity [a.u.]')
ax.set_title('Overlapping Spectra')

# Add a legend
custom_lines = [Line2D([0], [0], color=colors(i+1), lw=2) for i in range(num_spectra)]
legend_labels = ['Raw 1% Pt/Al2O3', '1% Pt/Al2O3 calcinated at 200C', '1% Pt/Al2O3 calcinated at 300C', '1% Pt/Al2O3 calcinated at 400C', '1% Pt/Al2O3 calcinated at 600C' ]
ax.legend(custom_lines, legend_labels)


# Show the plot
plt.show()
```
